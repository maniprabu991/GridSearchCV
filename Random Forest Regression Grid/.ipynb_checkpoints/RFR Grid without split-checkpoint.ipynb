```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first=True).astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
dependent = np.ravel(dependent)
param_grid = {
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'max_features': ['sqrt','log2',None],
    'n_estimators':[10,100]
}

grid = GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(independent,dependent)
```

```python
result = grid.cv_results_

print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(result)
```

```python
table
```

```python
Age_input=int(input("Age:"))
BMI_input=float(input("BMI:"))
Child_input=int(input("Children:"))
Gender_input=int(input("Sex Male 0 or 1:"))
Smoker_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[Age_input,BMI_input,Child_input,Gender_input,Smoker_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
